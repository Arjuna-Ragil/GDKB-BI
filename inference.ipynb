```python
import joblib
import numpy as np
from data import data_setup
from prescriptive import prescriptive_analysis
```

```python
# USER INPUT
data_path = 'dataset/transaksi_bulan_ini_dummy.csv'
restock_budget = 1000000 
inventory_capacity = 10000
profit_margin = 0.3
```

```python
# DATA PREPARATION & PREDICTIVE
df_prediction, sku_list, unit_price = data_setup(data_path, 'context/training_col.pkl')

model = joblib.load('model/model_final.pkl')
prediction_result = model.predict(df_prediction)

next_month_demand = {sku: int(np.round(pred)) for sku, pred in zip(sku_list, prediction_result)}

print("Demand Prediction")
for sku, qty in next_month_demand.items():
    print(f"{sku} : {qty} unit")
print("\nNew Unit Price Average:")
for sku, harga in unit_price.items():
    print(f"{sku} : Rp {harga:,.2f}")
```

```python
# PRESCRIPTIVE
prescriptive_analysis(unit_price, next_month_demand, restock_budget, inventory_capacity, profit_margin)
```
